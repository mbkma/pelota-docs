# Pelota Coordinate System

Pelota uses a right-handed 3D coordinate system aligned with Godot's defaults.

## Axes

- x: left and right across the court
- y: vertical height above the ground
- z: depth along the court from baseline to baseline

The court is centered at the origin, so the net sits at z = 0 and the baselines sit near +/- 13 meters.
